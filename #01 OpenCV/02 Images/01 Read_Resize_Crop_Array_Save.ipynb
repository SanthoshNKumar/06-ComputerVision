### Save Image from camera

```python
import cv2
from PIL import Image
import os
from tensorflow.keras.preprocessing.image import load_img

# Define the path
directory = "C://data"

# Open the camera
video = cv2.VideoCapture(0)

# Read the frames from the video
_, frame = video.read()

# RGB conversion
im = Image.fromarray(frame, 'RGB')

# Save in the path
im.save(os.path.join(directory,"2.jpg"), "JPEG")

# Release the Web CAM
video.release()

# Destroy the Web cam sub windows
cv2.destroyAllWindows()
```

### Resize Image

```python
import cv2

# Original image
path = "image1.jpg"
img  = cv2.imread(path)

# resize
width ,height = 1000 , 1000
imgResize = cv2.resize(img,(width,height))

# Crop
imgCropped = img[100:240,130:280]

# Croped Resize
imCropResize  = cv2.resize(imgCropped,(img.shape[1],img.shape[0]))

import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2,figsize=(10,10))
fig.suptitle('Crop and Resize Image')
ax[0,0].imshow(img)
ax[0,0].set_title("Original Image")

ax[0,1].imshow(imgResize)
ax[0,1].set_title("Resize Image")

ax[1,0].imshow(imgCropped)
ax[1,0].set_title("Croped Image")

ax[1,1].imshow(imCropResize)
ax[1,1].set_title("Crop Resized Image")
```
