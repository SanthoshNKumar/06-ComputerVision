### Save Image from camera

```python
import cv2
from PIL import Image
import os
from tensorflow.keras.preprocessing.image import load_img

# Define the path
directory = "C://data"

# Open the camera
video = cv2.VideoCapture(0)

# Read the frames from the video
_, frame = video.read()

# RGB conversion
im = Image.fromarray(frame, 'RGB')

# Save in the path
im.save(os.path.join(directory,"2.jpg"), "JPEG")

# Release the Web CAM
video.release()

# Destroy the Web cam sub windows
cv2.destroyAllWindows()
```

```python
import cv2

# Original image
path = "image1.jpg"
img  = cv2.imread(path)

# resize
width ,height = 1000 , 1000
imgResize = cv2.resize(img,(width,height))

# Crop
imgCropped = img[100:240,130:280]

# Croped Resize
imCropResize  = cv2.resize(imgCropped,(img.shape[1],img.shape[0]))

import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2,figsize=(10,10))
fig.suptitle('Crop and Resize Image')
ax[0,0].imshow(img)
ax[0,0].set_title("Original Image")

ax[0,1].imshow(imgResize)
ax[0,1].set_title("Resize Image")

ax[1,0].imshow(imgCropped)
ax[1,0].set_title("Croped Image")

ax[1,1].imshow(imCropResize)
ax[1,1].set_title("Crop Resized Image")
```

### Appending 

```python
import numpy
import cv2 
 
img = cv2.imread('image1.jpg') 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
 
## won't work
##horizontalAppendedIGrayImg = numpy.hstack((img,gray))
 
grayImageBGRspace = cv2.cvtColor(gray,cv2.COLOR_GRAY2BGR)
 
horizontalAppendedIGrayImg = numpy.hstack((img, grayImageBGRspace))

verticalAppendedIGrayImg = numpy.vstack((img, grayImageBGRspace))
 
#cv2.imshow('Horizontal Appended Gray Img', horizontalAppendedIGrayImg)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(5,5))
fig.suptitle('Measure the Distance of Object')
ax1.imshow(horizontalAppendedIGrayImg)
ax1.axis("off")

ax2.imshow(verticalAppendedIGrayImg)
ax2.axis("off")
```

### Flipping an image

```python
import cv2
  
originalImage = cv2.imread('image1.jpg')
  
flipVertical = cv2.flip(originalImage, 0)
flipHorizontal = cv2.flip(originalImage, 1)
flipBoth = cv2.flip(originalImage, -1)
 
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2,figsize=(10,10))
fig.suptitle('Crop and Resize Image')

ax[0,0].imshow(originalImage,cmap='gray')
ax[0,0].set_title("Original Image")

ax[0,1].imshow(flipVertical)
ax[0,1].set_title("Flip verticle")

ax[1,0].imshow(flipHorizontal)
ax[1,0].set_title("Flip Horizontal")

ax[1,1].imshow(flipBoth)
ax[1,1].set_title("Flip Both")
```
