```python
import cv2

# reading the image
img = cv2.imread('convey_belt.jpg', 1)

belt = img[131:259,92:333] # the ROI Area

grey_belt = cv2.cvtColor(belt,cv2.COLOR_BGR2GRAY)

# Apply threshold
_,threshold = cv2.threshold(grey_belt,80,255,cv2.THRESH_BINARY)

# Detect Nuts using 'contours'
contours,_ = cv2.findContours(threshold,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    (x,y,w,h) = cv2.boundingRect(cnt)
    
    # Draw Rectangle
    cv2.rectangle(belt,(x,y),(x+w,y+h),(0,255,0),3)

print(len(contours))

# displaying the image
cv2.imshow('belt', belt)

# displaying the image
cv2.imshow('grey_belt', grey_belt)

# displaying the image
cv2.imshow('threshold_belt', threshold)

# wait for a key to be pressed to exit
cv2.waitKey(0)

# close the window
cv2.destroyAllWindows()
```
