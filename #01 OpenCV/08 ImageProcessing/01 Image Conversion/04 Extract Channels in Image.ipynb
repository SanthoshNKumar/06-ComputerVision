### Extract Red Channel from Color Image

```python
import cv2
 
#read image
src = cv2.imread('image2.jpg', cv2.IMREAD_UNCHANGED)
print(src.shape)

#extract red channel
red_channel = src[:,:,2]

#write red channel to greyscale image
cv2.imwrite('cv2-red-channel.png',red_channel) 
```

```python
import matplotlib.pyplot as plt

no_blue_chl=plt.imread("cv2-red-channel.png")

plt.imshow(no_blue_chl)
```

```python
import cv2
import numpy as np

#read image
src = cv2.imread('image2.jpg', cv2.IMREAD_UNCHANGED)
print(src.shape)

# extract red channel
red_channel = src[:,:,2]

# create empty image with same shape as that of src image
red_img = np.zeros(src.shape)

#assign the red channel of src to empty image
red_img[:,:,2] = red_channel

#save image
cv2.imwrite('cv2-red-channel.png',red_img) 
```

```python
import matplotlib.pyplot as plt

no_blue_chl=plt.imread("cv2-red-channel.png")

plt.imshow(no_blue_chl)
```

### Get Blue Channel from Image

```python
import cv2
import numpy as np

#read image
src = cv2.imread('image2.jpg', cv2.IMREAD_UNCHANGED)
print(src.shape)

# extract blue channel
blue_channel = src[:,:,0]

# create empty image with same shape as that of src image
blue_img = np.zeros(src.shape)

#assign the red channel of src to empty image
blue_img[:,:,0] = blue_channel

#save image
cv2.imwrite('cv2-blue-channel.png',blue_img) 
```

```python
import matplotlib.pyplot as plt

blue_chl=plt.imread("cv2-blue-channel.png")

plt.imshow(blue_chl)
```

### Remove Red Channel from Image

```python
import cv2
import numpy as np

#read image
src = cv2.imread('image2.jpg', cv2.IMREAD_UNCHANGED)
print(src.shape)

# assign red channel to zeros
src[:,:,2] = np.zeros([src.shape[0], src.shape[1]])

#save image
cv2.imwrite('no-red-channel.png',src) 
```

```python
import matplotlib.pyplot as plt

red_rmd_chl=plt.imread("no-red-channel.png")

plt.imshow(red_rmd_chl)
```
