```python
# RGB :Red Green Blue
# BGR : Blue Green Red

An RGB color is stored in a structure 
    - Blue occupying the least significant “area”
    - Green the second least
    - Red the third least
    
BGR is the same, except the order of areas is reversed
    - Red occupies the least significant area
    - Green the second (still)
    - Blue the third
    
    The main difference between RGB versus BGR is the arrangement of the subpixels for Red, Green, and Blue
    
```

```python
HSV Color Model : 
    
HSV – (hue, saturation, value), also known as HSB (hue, saturation, brightness)

Hue:
    - It represents color. 
    - It is an angle from 0 degrees to 360 degrees.
    -   Angle	Color
         0-60	Red
         60-120	Yellow
         120-180	Green
         180-240	Cyan
         240-300	Blue
         300-360	Magenta

Saturation: 
    - It indicates the range of grey in the color space.
    - It ranges from 0 to 100%. Sometimes the value is calculated from 0 to 1. 
    - When the value is '0,' the color is grey and when the value is '1,' the color is a primary color.
    
Value 
    - It is the brightness of the color and varies with color saturation
    - It ranges from 0 to 100%. 
    - When the value is '0' the color space will be totally black. 
    - With the increase in the value, the color space brightness up and shows various colors.
    
An HSV is another type of color space in which H stands for Hue, S stands for Saturation and V stands for Value.
```

```python
import cv2

img = cv2.imread("image1.jpg",1)

img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

img_hsv = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)

print("Shape of RGB:",img_bgr.shape)
print("Shape of Gray:",img_gray.shape)
print("Shape of HSV:",img_hsv.shape)

import matplotlib.pyplot as plt

fig, (ax1, ax2,ax3) = plt.subplots(1, 3,figsize=(15,7))
fig.suptitle('RGB,Gray, HSV Conversion')
ax1.imshow(img_bgr)
ax1.set_title("RGB")
ax1.axis('off')

ax2.imshow(img_gray,cmap='gray')
ax2.set_title("Gray")
ax2.axis('off')

ax3.imshow(img_hsv)
ax3.set_title("HSV")
ax3.axis('off')
```
