```python
1. Read the Image
2. Convert to Gray Scale
3. Apply 'threshold' function considering the values 127 and 255
```

```python
# import opencv
import cv2
 
# Load the input image
img = cv2.imread("image1.jpg",1)

# Original Image
original = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# grayscale the image
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# define a threshold, 128 is the middle of black and white in grey scale
thresh = 128
 
# Apply Thresold function
(thresh, blackAndWhiteImage) = cv2.threshold(gray_image, thresh, 255, cv2.THRESH_BINARY)

print("Shape of Original Image:",original.shape)  

print("Shape of Grayscale Image:",gray_image.shape) 

print("Shape of Binary Image:",blackAndWhiteImage.shape)  

import matplotlib.pyplot as plt

fig, (ax1, ax2,ax3) = plt.subplots(1, 3,figsize=(10,5))
fig.suptitle('RGB,GrayScale and Binary Image')

ax1.imshow(original,cmap='gray')
ax1.set_title("Original Image")
ax1.axis('off')

ax2.imshow(gray_image,cmap='gray')
ax2.set_title("GrayScale")
ax2.axis('off')

ax3.imshow(blackAndWhiteImage,cmap='gray')
ax3.set_title("Original Image(Binary)")
ax3.axis('off')
```
