```python
import numpy as np
import cv2 as cv

img = cv.imread('image1.jpg',0) # getting Gray Sale image

edges = cv.Canny(img,100,200) # (Image,threshold1,threshold2)

from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))

plt.subplot(121)
plt.imshow(img,cmap = 'gray')
plt.title('Original Image')
plt.axis("off")

plt.subplot(122)
plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image')
plt.axis("off")
plt.show()
```
