### Method : using Open CV 'bitwise_and'

```python
import cv2

# load the image and display it
image = cv2.imread("img/1.jpg")

# convert the image to grayscale and blur it slightly
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blurred = cv2.GaussianBlur(gray, (7, 7), 0)

# using normal thresholding (rather than inverse thresholding)
(T, thresh) = cv2.threshold(blurred, 200, 255, cv2.THRESH_BINARY)

(T, threshInv) = cv2.threshold(blurred, 200, 255,cv2.THRESH_BINARY_INV)

masked = cv2.bitwise_and(image, image, mask=threshInv)

import matplotlib.pyplot as plt

fig,ax = plt.subplots(2, 3,figsize=(15,5))
fig.suptitle('Image Masking using Threshold')

ax[0,0].imshow(image,cmap="gray")
ax[0,0].axis('off')
ax[0,0].set_title("Original Image")

ax[0,1].imshow(gray,cmap="gray")
ax[0,1].axis('off')
ax[0,1].set_title("GrayScale")

ax[0,2].imshow(blurred,cmap="gray")
ax[0,2].axis('off')
ax[0,2].set_title("Blured")

ax[1,0].imshow(thresh,cmap="gray")
ax[1,0].axis('off')
ax[1,0].set_title("Threshold Binary")

ax[1,1].imshow(threshInv,cmap="gray")
ax[1,1].axis('off')
ax[1,1].set_title("Threshold Binary Inverse")

ax[1,2].imshow(masked,cmap="gray")
ax[1,2].axis('off')
ax[1,2].set_title("Masked")
```

### Method 2

```python
import numpy as np
import cv2

# Read image
original_img = cv2.imread("img/1.jpg")

#gary scale image
grayscale_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

# Set threshold and maxValue
thresh = 210
maxValue = 255

_, binary = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY);

selection = np.zeros_like(original_img)

t = 0.8
binary_mask = binary < t

selection[binary_mask] = image[binary_mask]

fig, ax = plt.subplots()
plt.imshow(selection)
plt.axis("off")
plt.show()
```
