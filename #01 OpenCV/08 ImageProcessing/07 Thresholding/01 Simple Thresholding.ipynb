### Binary Thresholding (THRESH_BINARY)

```python
# import opencv
import cv2

# Read image
original_img = cv2.imread("Thresholding01.JPG")

#gary scale image
grayscale_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

# Set threshold and maxValue
thresh = 0
maxValue = 255 

# Basic threshold example
th1, dst1 = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY);

# Set threshold and maxValue
thresh = 127
maxValue = 255

# Basic threshold example
th2, dst2 = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY);

# Set threshold and maxValue
thresh = 127
maxValue = 128
# Basic threshold example
th3, dst3 = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY);

import matplotlib.pyplot as plt

fig, (ax1, ax2,ax3,ax4) = plt.subplots(1, 4,figsize=(20,4))
fig.suptitle('Simple_Threshold_type_THRESH_BINARY')

ax1.imshow(grayscale_img,cmap="gray") 
ax1.axis("off")
ax1.set_title("Gray")

ax2.imshow(dst1,cmap="gray")
ax2.axis("off")
ax2.set_title("Binary(thresh =0 and maxValue = 255)")

ax3.imshow(dst2,cmap="gray")
ax3.axis("off")
ax3.set_title("Binary(thresh =127 and maxValue = 255)")

ax4.imshow(dst3,cmap="gray")
ax4.axis("off")
ax4.set_title("Binary(thresh =127 and maxValue = 128)")
```

### Inverse-Binary Thresholding ( THRESH_BINARY_INV )

```python
# import opencv
import cv2

# Read image
original_img = cv2.imread("Thresholding01.JPG")

#gary scale image
grayscale_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

# *****************************
# Set threshold and maxValue
thresh = 0
maxValue = 255 

# Basic threshold example
th1, dst1 = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY_INV);


# *****************************
# Set threshold and maxValue
thresh = 127
maxValue = 255

# Basic threshold example
th2, dst2 = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY_INV);


# *****************************
# Set threshold and maxValue
thresh = 127
maxValue = 128
# Basic threshold example
th3, dst3 = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY_INV);

import matplotlib.pyplot as plt

fig, (ax1, ax2,ax3,ax4) = plt.subplots(1, 4,figsize=(20,4))
fig.suptitle('Simple_Threshold_type_THRESH_BINARY')

ax1.imshow(grayscale_img,cmap="gray") 
ax1.axis("off")
ax1.set_title("Gray")

ax2.imshow(dst1,cmap="gray")
ax2.axis("off")
ax2.set_title("Binary(thresh =0 and maxValue = 255)")

ax3.imshow(dst2,cmap="gray")
ax3.axis("off")
ax3.set_title("Binary(thresh =127 and maxValue = 255)")

ax4.imshow(dst3,cmap="gray")
ax4.axis("off")
ax4.set_title("Binary(thresh =127 and maxValue = 128)")
```

```python
import cv2

# Read image
original_img = cv2.imread("1.jpg")

#gary scale image
grayscale_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

# Set threshold and maxValue
thresh = 210
maxValue = 255

_, binary = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY);

# Basic threshold example
_, binary_inv = cv2.threshold(grayscale_img, thresh, maxValue, cv2.THRESH_BINARY_INV);

import matplotlib.pyplot as plt

fig, (ax1, ax2,ax3) = plt.subplots(1, 3,figsize=(15,4))
fig.suptitle('BInary and Inverse Binary Conversion')

ax1.imshow(original_img,cmap="gray")
ax1.axis("off")
ax1.set_title("Original")

ax2.imshow(binary,cmap="gray")
ax2.axis("off")
ax2.set_title("Binary")

ax3.imshow(binary_inv,cmap="gray")
ax3.axis("off")
ax3.set_title("Inverse Binary")
```
