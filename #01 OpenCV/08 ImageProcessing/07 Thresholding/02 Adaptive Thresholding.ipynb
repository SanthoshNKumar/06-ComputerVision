```python
# Python program to illustrate 
# adaptive thresholding type on an image
       
# organizing imports 
import cv2 
import numpy as np 
   
# path to input image is specified and  
# image is loaded with imread command 
image1 = cv2.imread('page.jpg') 
   
# cv2.cvtColor is applied over the
# image input with applied parameters
# to convert the image in grayscale 
img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

# Apply threshold
_,threshold = cv2.threshold(img,80,255,cv2.THRESH_BINARY)
   
# applying different thresholding 
# techniques on the input image
thresh1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                          cv2.THRESH_BINARY, 199, 5)
  
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 199, 20)

import matplotlib.pyplot as plt

fig, (ax1, ax2,ax3,ax4) = plt.subplots(1, 4,figsize=(15,4))
fig.suptitle('Measure the Distance of Object')
ax1.imshow(image1,cmap='gray')
ax1.axis("Off")
ax1.set_title("Original")

ax2.imshow(threshold,cmap='gray')
ax2.axis("Off")
ax2.set_title("Simple Threshold")

ax3.imshow(thresh1,cmap='gray')
ax3.axis("Off")
ax3.set_title("Adaptive mean")

ax4.imshow(thresh2,cmap='gray')
ax4.axis("Off")
ax4.set_title("Adaptive Gaussian")
```
