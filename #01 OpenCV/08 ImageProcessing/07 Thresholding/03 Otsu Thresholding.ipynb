```python
'''
A problem with simple thresholding is that you have to manually specify the threshold value.
Otsuâ€™s method then computes an optimal threshold value T such that the variance between the background and foreground peaks 
is minimal.
```

```python
import cv2

# load the image and display it
image = cv2.imread("img/1.jpg")

# convert the image to grayscale and blur it slightly
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blurred = cv2.GaussianBlur(gray, (7, 7), 0)

# apply Otsu's automatic thresholding which automatically determines
# the best threshold value
(T, threshInv) = cv2.threshold(blurred, 0, 255,cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)

print("[INFO] otsu's thresholding value: {}".format(T))

# visualize only the masked regions in the image
masked = cv2.bitwise_and(image, image, mask=threshInv)

import matplotlib.pyplot as plt

fig,ax = plt.subplots(2, 3,figsize=(15,5))
fig.suptitle('Image Masking using Threshold')

ax[0,0].imshow(image,cmap="gray")
ax[0,0].axis('off')
ax[0,0].set_title("Original Image")

ax[0,1].imshow(gray,cmap="gray")
ax[0,1].axis('off')
ax[0,1].set_title("GrayScale")

ax[0,2].imshow(blurred,cmap="gray")
ax[0,2].axis('off')
ax[0,2].set_title("Blured")

ax[1,0].imshow(threshInv,cmap="gray")
ax[1,0].axis('off')
ax[1,0].set_title("Threshold Binary Inverse")

ax[1,1].imshow(masked,cmap="gray")
ax[1,1].axis('off')
ax[1,1].set_title("Masked")
```
