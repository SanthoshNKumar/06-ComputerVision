### Averaging

```python
import cv2 as cv
import numpy as np


img = cv.imread('image1.jpg')
blur = cv.blur(img,(5,5))

from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))

plt.subplot(121),plt.imshow(img,cmap="gray"),plt.title('Original')
plt.axis("off")

plt.subplot(122),plt.imshow(blur,cmap="gray"),plt.title('Blurred')
plt.axis("off")
plt.show()
```

### Gaussian Blurring

```python
import cv2 as cv
import numpy as np


img = cv.imread('image1.jpg')
blur = cv.GaussianBlur(img,(5,5),0)

from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))

plt.subplot(121),plt.imshow(img,cmap="gray"),plt.title('Original')
plt.axis("off")

plt.subplot(122),plt.imshow(blur,cmap="gray"),plt.title('Gaussian')
plt.axis("off")
plt.show()
```

### Median Blurring

```python
import cv2 as cv
import numpy as np


img = cv.imread('image1.jpg')
median = cv.medianBlur(img,5)

from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))

plt.subplot(121),plt.imshow(img,cmap="gray"),plt.title('Original')
plt.axis("off")

plt.subplot(122),plt.imshow(median,cmap="gray"),plt.title('Median')
plt.axis("off")
plt.show()
```

### Bilateral Filtering

```python
import cv2 as cv
import numpy as np


img = cv.imread('image1.jpg')
blur = cv.bilateralFilter(img,9,75,75)

from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))

plt.subplot(121),plt.imshow(img,cmap="gray"),plt.title('Original')
plt.axis("off")

plt.subplot(122),plt.imshow(blur,cmap="gray"),plt.title('Bilateral')
plt.axis("off")
plt.show()
```
