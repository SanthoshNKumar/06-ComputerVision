```python
'''

Morphological Filters:
    - The idea of the morphological filter are shrink and let grow process

Shirnk:
    - The word “shrink” means using median filter to round off the large structures and to remove the small structures 

Let Grow:
    -  Remaining structures are grow back by the same amount.
    
- Morphological Filters
    - Erosion
    - Dilation
    - Opening
    - Closing
    - Morphological Gradient
    - Top hat/white hat and black hat


Erosion : Shrink Process
    - Erosion removes pixels on object boundaries
    - It removes the anamolies

Dilation: Grow process
    - Dilation adds pixels to the boundaries of objects in an image
    - Dilation is A XOR B.
    - It Fills the holes and broken ares
    
Opening:
    - An opening is an erosion followed by a dilation. It’s effective to remove salt noise

Closing:
    - Opposite to an opening,A closing is a dilation followed by an erosion. It’s effective to remove pepper noise
    
Morphological Gradeint:
    - A morphological gradient is the difference between a dilation and erosion.
    
Top hat/white hat and black hat:
    - A top hat (also known as a white hat) morphological operation.
    - It is the difference between the original (grayscale/single channel) input image and the opening
```

#### Erosion
    - Erosion removes pixels on object boundaries
#### Dilation
    - Dilation adds pixels to the boundaries of objects in an image

```python
# dilation of images.
import cv2
import numpy as np
 
# Reading the input image
img = cv2.imread('text_blackImg.png', 0) # get the Gray scale image
 
# Taking a matrix of size 5 as the kernel
kernel = np.ones((5,5), np.uint8)
 
# (img,kernel,iterations)
img_erosion = cv2.erode(img, kernel, iterations=1) # Shrink
img_dilation = cv2.dilate(img, kernel, iterations=1) # Grow 


import matplotlib.pyplot as plt

f, ax = plt.subplots(nrows=1,ncols=3,figsize=(15,10))

plt.sca(ax[0]); 
plt.imshow(img,cmap="gray"); 
plt.title("Grayscale")
plt.axis("off")

plt.sca(ax[1]); 
plt.imshow(img_erosion,cmap="gray"); 
plt.title("Erosion")
plt.axis("off")

plt.sca(ax[2]); 
plt.imshow(img_dilation,cmap="gray"); 
plt.title("Dilation")
plt.axis("off")

plt.show()
```

### Opening 
        - An opening is an erosion followed by a dilation. It’s effective to remove salt noise

```python
import cv2
import numpy as np
 
# Reading the input image
gray_img = cv2.imread('text_blackImg.png', 0) # get the Gray scale image
 
kernelSizes = [(3, 3), (5, 5), (7, 7)]

import matplotlib.pyplot as plt

f, ax = plt.subplots(nrows=1,ncols=3,figsize=(15,10))

i= -1

# loop over the kernels sizes
for kernelSize in kernelSizes:
    i= i+1
    
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    opening = cv2.morphologyEx(gray_img, cv2.MORPH_OPEN, kernel)
    
    plt.sca(ax[i]); 
    plt.imshow(opening,cmap="gray"); 
    plt.title("Opening_Kernel_"+str(i))
    plt.axis("off")
    
```

### Morphological gradient

```python
import cv2
import numpy as np
 
# Reading the input image
gray_img = cv2.imread('text_blackImg.png', 0) # get the Gray scale image
 
kernelSizes = [(3, 3), (5, 5), (7, 7)]

import matplotlib.pyplot as plt

f, ax = plt.subplots(nrows=1,ncols=3,figsize=(15,10))

i= -1

# loop over the kernels sizes
for kernelSize in kernelSizes:
    i= i+1
    
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    gradient = cv2.morphologyEx(gray_img, cv2.MORPH_GRADIENT, kernel)
    
    plt.sca(ax[i]); 
    plt.imshow(gradient,cmap="gray"); 
    plt.title("Closing_Kernel_"+str(i))
    plt.axis("off")
    
```

## Top hat/white hat and black hat
    -  A top hat (also known as a white hat) morphological operation.

```python
import cv2
import numpy as np
 
# Reading the input image
gray_img = cv2.imread('bike.jpg', 0) # get the Gray scale image

# construct a rectangular kernel (13x5) and apply a blackhat
# operation which enables us to find dark regions on a light
# background
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (13, 5))
blackhat = cv2.morphologyEx(gray_img, cv2.MORPH_BLACKHAT, rectKernel)

# similarly, a tophat (also called a "whitehat") operation will
# enable us to find light regions on a dark background
tophat = cv2.morphologyEx(gray_img, cv2.MORPH_TOPHAT, rectKernel)

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10,5))
fig.suptitle('Black and Top Hat')
ax1.imshow(blackhat)
ax1.set_title("Black hat")
ax1.axis("off")

ax2.imshow(tophat)
ax2.set_title("Top Hat")
ax2.axis("off")
```
