```python
import cv2

img = cv2.imread("image1.jpg",1)

#
gray = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

cv2.circle(gray,(193,130), 10, (0,0,255), -1)  # (Image, center point,RediusOfCircle,color,thickness)

cv2.circle(gray,(202,267), 20, (0,0,255), 3)  # (Image, center point,RediusOfCircle,color,thickness)

print(gray.shape)

import matplotlib.pyplot as plt
plt.figure(figsize = (10,7))
plt.axis("off")
plt.imshow(gray)
plt.show()
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
    
    cv2.circle(gray,(447,63), 63, (0,0,255), -1)

    # Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
