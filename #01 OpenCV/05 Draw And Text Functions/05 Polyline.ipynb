```python
import numpy as np  
import cv2 

img = cv2.imread(r"image1.jpg",1)  

gray = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

#defining points for polylines   # Get this Coordinates Values using Paint
pts = np.array([[36,48],[168,34],[319,218],[98,347]], np.int32)  

cv2.polylines(gray, [pts], True, (0,255,0), 2)  

import matplotlib.pyplot as plt
plt.figure(figsize = (10,7))
plt.axis("off")
plt.imshow(gray)
plt.show()
```
