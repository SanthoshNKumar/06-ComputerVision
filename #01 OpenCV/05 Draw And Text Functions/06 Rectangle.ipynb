```python
import numpy as np  
import cv2 

img = cv2.imread(r"image1.jpg",1)  

gray = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# (img,(xmin,ymin),(xmax,ymax),color,thickness)
gray = cv2.rectangle(gray,(164,109),(228,158),(0,255,0),2)

gray = cv2.rectangle(gray,(130,246),(258,337),(255,0,0),-1) # Fill color 

import matplotlib.pyplot as plt
plt.figure(figsize = (10,7))
plt.axis("off")
plt.imshow(gray)
plt.show()
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    #gray = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    cv2.rectangle(gray,(384,0),(510,128),(0,255,0),3) # (img,Point1,point2,color,thikness)

    # Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
