```python
import cv2

img = cv2.imread("image1.jpg",1)

gray = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

cv2.arrowedLine(gray,(43,63),(368,71),(0,0,255),5) # (img,startpoint(sharp),endpoint, color,thickness)
cv2.arrowedLine(gray,(301,133),(317,363),(255,0,0),5) #

print(gray.shape)

import matplotlib.pyplot as plt
plt.figure(figsize = (10,7))
plt.axis("off")
plt.imshow(gray)
plt.show()
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
    
    #cv2.line(gray,(0,0),(511,511),(255,0,0),5) # (image,point1,point2,Color,Thickness)
    cv2.line(gray,(0,100),(700,100),(0,0,255),5) #
    cv2.line(gray,(0,200),(700,200),(255,0,0),5) #

    # Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
