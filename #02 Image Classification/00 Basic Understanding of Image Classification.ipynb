```python
'''
AlexNet
    - Winner of 2012 ImageNet Competition
    - AlexNet is basically used for Object detection task
    - It consists of 8 Layers (5 Conv Layer and 3 FC layer)
    - AlexNet was trained on ImageNet
    - Input size of the image for AlexNet is 227*227*3
    
Other Features in AlexNet:
    - ReLU Nonlinearity
        - AlexNet uses Rectified Linear Units (ReLU) instead of the tanh function, which was standard at the time.
        - With CIFAR-10 dataset six times faster (Relu) than a CNN using tanh.
    - Multiple GPUs
        - Back in the day, GPUs were still rolling around with 3 gigabytes of memory.
        - This was especially bad because the training set had 1.2 million images
        - AlexNet allows for multi-GPU training by putting half of the model’s neurons on one GPU and the other 
          half on another GPU.
    - dropout
        - Use dropout instead of regularisation to deal with overfitting. However, the training time is doubled with the 
          dropout rate of 0.5
    - Overlap pooling
        - Overlap pooling to reduce the size of the network. It reduces the top-1 and top-5 error rates by 0.4% and 0.3%, 
          respectively
          
AlexNet Training and Results:
    - The model uses a stochastic gradient descent optimization function with batch size,
      momentum, and weight decay set to 128, 0.9, and 0.0005 respectively. 
      All the layers use an equal learning rate of 0.001. 
      To address overfitting during training, AlexNet uses both data augmentation and dropout layers. 
      It took approximately six days to train on two GTX 580 3GB GPUs for 90 cycles.
```

```python
'''
Auxilary Classifier?
    - Use this to handle the issue of Vanishing Gradient problem
    - The auxiliary classifier improves the convergence of the network. 
      They argue that it can help reduce the effect of the vanishing gradient problem in the deep networks by pushing the 
      useful gradient to earlier layers (to reduce the loss). 
    - They are classifier heads we attach to layers before the end of the network.
    - The motivation is to push useful gradients to the lower layers to make them immediately useful 
      and improve the convergence during training by combatting the vanishing gradient problem.
    - # The total loss used by the inception net during training.
      total_loss = real_loss + 0.3 * aux_loss_1 + 0.3 * aux_loss_2
```

```python
'''
Inception V2:
    - Inception v2 and Inception v3 were presented in the same paper
    - The authors proposed a number of upgrades which increased the accuracy and reduced the computational complexity. 
    - Reduce representational bottleneck.
    - Using smart factorization methods, convolutions can be made more efficient in terms of computational complexity
    
Solution:
    - Factorize 5x5 convolution to two 3x3 convolution operations to improve computational speed.
        - A 5x5 convolution is 2.78 times more expensive than a 3x3 convolution
        - Stacking two 3x3 convolutions infact leads to a boost in performance

Note: a 5x5 convolution is 2.78 times more expensive than a 3x3
      So stacking two 3x3 convolutions infact leads to a boost in performance
```

```python
'''
Contd.. Solution 

   - factorize convolutions of filter size nxn to a combination of 1xn and nx1 convolutions. 
     For example, a 3x3 convolution is equivalent to first performing a 1x3 convolution, and then performing a 3x1 
     convolution on its output. found this method to be 33% more cheaper than the single 3x3 convolution
```

```python
'''
Contd.. Solution 
    - Making the inception more wider instead deeper
    - The filter banks in the module were expanded (made wider instead of deeper) to remove the representational bottleneck.
    - If the module was made deeper instead, there would be excessive reduction in dimensions, and hence loss of information.
```

```python
'''
Difference between Inception v2 and V3
    - Inception Net v3 incorporated all of the above upgrades stated for Inception v2
        - RMSProp Optimizer.
        - Factorized 7x7 convolutions.
        - BatchNorm in the Auxillary Classifiers.
        - Label Smoothing.
```

```python
'''
ResNet:(Residual Network)
    - Won 1st place in the ILSVRC 2015 classification competition
    - ResNet, which was proposed in 2015 by researchers at Microsoft Research introduced a new architecture
      called Residual Network.
    - Varinats of ResNet (18,34,50,101,151)
    - Main Objective is to solve the Vanishing Gradient Problem
    
Why RestNet Introduced?
    
    - In the old Models as we go deeper(increasing the number of layer in th network) the error is increasing
      (Both Training and Test) and main reason is vanishing and explofding Gradient descent.
    - After the first CNN-based architecture (AlexNet) that win the ImageNet 2012 competition, Every subsequent 
      winning architecture uses more layers in a deep neural network to reduce the error rate
    - This(AlexNet) works for less number of layers, but when we increase the number of layers, there is a common problem in 
      deep learning associated with that called Vanishing/Exploding gradient.
    - This causes the gradient to become 0 or too large.
    - Thus when we increases number of layers, the training and test error rate also increases.
    - However, increasing network depth does not work by simply stacking layers together. Deep networks are hard to train 
      because of the vanishing gradient problem 
    - Before ResNet, there had been several ways to deal the vanishing gradient issue,adds an auxiliary loss in a middle 
      layer as extra supervision, but none seemed to really tackle the problem once and for all.
    - The core idea of ResNet is introducing a so-called “identity shortcut connection” that skips one or more layers
    
Need of ResNet?
    - Mostly in order to solve a complex problem, we stack some additional layers in the Deep Neural Networks which results 
      in improved accuracy and performance. The intuition behind adding more layers is that these layers progressively 
      learn more complex features.
    - Ex: For example, in case of recognising images, the first layer may learn to detect edges, the second layer may 
      learn to identify textures and similarly the third layer can learn to detect objects and so on. But it has been found 
      that there is a maximum threshold for depth with the traditional Convolutional neural network model. 
    - This suggests that with adding more layers on top of a network, its performance degrades
    
How ResNet Works?
    - In deep learning it is always hard to train becasue of vanishing Gradient problem.ResNet introuce something called 
      skip connection ("Add the original input to the output of the convolution block").
      In tradistional CNN, Conovolution layer are stacked one after the other. with resNet colvolution layer are stacked 
      like tradistion layer,over here now we also added original input to the output of the convloution Block(Skip Connection)
    - In order to solve the problem of the vanishing/exploding gradient,this architecture introduced the concept called 
      Residual Network. 
    - In this network we use a technique called skip connections . The skip connection skips training from a few 
      layers and connects directly to the output.
    - The advantage of adding this type of skip connection is because if any layer hurt the performance of architecture 
      then it will be skipped by regularization.So, this results in training very deep neural network without the problems 
      caused by vanishing/exploding gradient.
    - The skip connections in ResNet solve the problem of vanishing gradient in deep neural networks by allowing this 
      alternate shortcut path for the gradient to flow through.
```
