```python
from mtcnn.mtcnn import MTCNN
import cv2
```

```python
image = cv2.imread("/content/sample_data/dwayne-the-rock-.jpg")
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
detector = MTCNN()

faces = detector.detect_faces(image)
for face in faces:
  print(face)
```

```python
def create_box(image):
  faces = detector.detect_faces(image)
  bounidng_box = faces[0]['box']
  keypoints = faces[0]['keypoints']

  cv2.rectangle(image,(bounidng_box[0],bounidng_box[1]),
                (bounidng_box[0]+bounidng_box[2],bounidng_box[1]+ bounidng_box[3]),(0,155,255),3)
  
  cv2.circle(image,(keypoints['left_eye']),2,(0,155,255),2)
  cv2.circle(image,(keypoints['right_eye']),2,(0,155,255),2)
  cv2.circle(image,(keypoints['nose']),2,(0,155,255),2)
  cv2.circle(image,(keypoints['mouth_left']),2,(0,155,255),2)
  cv2.circle(image,(keypoints['mouth_right']),2,(0,155,255),2)

  return image
```

```python
maked_image = create_box(image)
```

```python
cv2_imshow(maked_image)
```

```python
# Read the Image
image = cv2.imread("/content/sample_data/4.jpg")

# Create Bounding Box
maked_image = create_box(image)

# Display the image outome
cv2_imshow(maked_image)
```
